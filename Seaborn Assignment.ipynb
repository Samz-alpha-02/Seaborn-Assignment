```python
#Q1
Seaborn is a popular data visualization library in Python that is built on top of Matplotlib. It provides 
a high-level interface for creating aesthetically pleasing and informative statistical graphics. Here are
five plots that can be created using Seaborn along with their uses:

1.Histograms (distplot): A histogram represents the distribution of a single variable by dividing it into bins
                         and displaying the frequency or density of observations within each bin. It helps 
                         visualize the underlying data distribution and identify patterns or outliers.

2.Scatter Plots (scatterplot): Scatter plots are used to visualize the relationship between two continuous 
                               variables. Each data point is represented by a marker on the plot, and the 
                               position of the marker indicates the values of the two variables. Scatter plots
                               help identify patterns, trends,and correlations between variables.

3.Bar Plots (barplot): Bar plots are useful for visualizing the relationship between a categorical variable and a
                       numerical variable. They display the average value of the numerical variable for each category 
                       as a bar. Bar plots help compare values across different categories and highlight differences 
                       or trends.

4.Violin Plots (violinplot): Violin plots are used to visualize the distribution of a numerical variable across 
                             different categories. They display a combination of a box plot and a kernel density 
                             plot. Violin plots provide insights into the median, quartiles, spread, and shape of
                             the distribution for each category.

5.Heatmaps (heatmap): Heatmaps are graphical representations of data in the form of a matrix, where the values are 
                      represented by colors. Heatmaps are commonly used to visualize correlations between variables 
                      or to display patterns in large datasets. They are particularly effective for identifying 
                      clusters or patterns in complex datasets.
```

```python
#Q2
import seaborn as sb
import matplotlib.pyplot as plt

# Load the "fmri" dataset
data = sb.load_dataset("fmri")

# Create line plot with different events and regions
sb.lineplot(data=data, x="timepoint", y="signal", hue="event", style="region")

# Set x and y axis labels
plt.xlabel('Timepoint')
plt.ylabel('Signal')

# Show the plot
plt.show()
```

```python
#Q3
import seaborn as sb
import matplotlib.pyplot as plt

data = sb.load_dataset("titanic")

box1 = sb.boxplot(data=data,x='pclass',y='age')
plt.xlabel('Passesnger Class')
plt.ylabel('Age')
plt.show()
```

```python
box2 = sb.boxplot(data=data,x='pclass',y='fare')
plt.xlabel('Passesnger Class')
plt.ylabel('Fare')
plt.show()
```

```python
#Q4
import seaborn as sb
import matplotlib.pyplot as plt

data = sb.load_dataset("diamonds")
sb.displot(data=data, x='price', hue='cut')

plt.show()
```

```python
#Q5
import seaborn as sb
import matplotlib.pyplot as plt

data = sb.load_dataset("iris")
sb.pairplot(data=data,hue="species")
plt.show()
```

```python
#Q6
import seaborn as sb
import matplotlib.pyplot as plt

data = sb.load_dataset("flights")
pivot_data = data.pivot("month", "year", "passengers")

plt.figure(figsize=(10, 8))
sb.heatmap(pivot_data, annot=True, fmt="d", cmap="YlGnBu")

plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Count Heatmap")

plt.show()
```
